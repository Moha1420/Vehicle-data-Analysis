```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv('c:/Users/ybede/OneDrive/Desktop/Electric_Vehicle.csv')
print(df.head())
#Histogram of Model Year
plt.figure(figsize=(8, 6))
sns.histplot(df['Model Year'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()

# Pie chart of Electric Vehicle Type
type_counts = df['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('Distribution of Electric Vehicle Type')
plt.show()

# Scatter plot of Electric Range vs Base MSRP
plt.figure(figsize=(10, 6))
plt.scatter(df['Electric Range'], df['Base MSRP'], c='b', alpha=0.5)
plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')
plt.title('Electric Range vs Base MSRP')
plt.show()

# Pairplot of Electric Range, Base MSRP, and Model Year
sns.pairplot(df[['Electric Range', 'Base MSRP', 'Model Year']], diag_kind='kde')
plt.suptitle('Pairplot of Electric Range, Base MSRP, and Model Year', y=1.02)
plt.show()

# Pie chart of Clean Alternative Fuel Vehicle (CAFV) Eligibility
cafv_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(cafv_counts, labels=cafv_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('Distribution of CAFV Eligibility')
plt.show()

# Bar plot of State counts
# Bar plot of State counts
plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, hue=state_counts.index, palette="viridis", legend=False)
plt.title('Count of Electric Vehicles by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Calculate the counts of electric vehicles by state
state_counts = df['State'].value_counts().head(10)

# Bar plot of State counts
plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis")
plt.title('Count of Electric Vehicles by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```
